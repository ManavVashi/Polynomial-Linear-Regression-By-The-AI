```python
import numpy as np
```

```python
x= 7 * np.random.rand(100,1) - 2.8
y= 7 * np.random.rand(100,1) - 2.8

# y = (a0 * x^2) + (a1 * y^2) + (a2 * x) + (a3 * y) + (a4 * x *  y) + a5
# let,

a0=1
a1=1
a2=0.2
a3=0.2
a4=0.1
a5=2

z = (a0 * x**2) + (a1 * y**2) + (a2 * x) + (a3 * y) + (a4 * x *  y) + a5 + np.random.randn(100,1)

#creating dataset using poynomial fucnction
```

```python
print(x.shape)
print(y.shape)
print(z.shape)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({"x":x[:,0], "y":y[:,0], "z":z[:,0]})
df.head()

#constructing DataFrame
```

```python
import plotly.express as px
graph=px.scatter_3d(df,x="x",y="y",z="z")
graph.show()

#plot dataset in a graph using plotly
```

```python
new_x=np.array([x,y]).reshape(100,2)
print(new_x.shape)

#creating input columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(new_x,y,test_size=0.2,random_state=3)

#spliting dataset into training data and testing data
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import PolynomialFeatures
plr=PolynomialFeatures(degree=30,include_bias=True)

new_x_train=plr.fit_transform(x_train)
new_x_test=plr.fit_transform(x_test)

#freating polynomial feature from simple feature
```

```python
print(new_x_train.shape)
print(new_x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

#import LinearRegression class from sci-kit learn
```

```python
lr.fit(new_x_train,y_train)

#training model
```

```python
y_predict=lr.predict(new_x_test)
y_predict

#teating model
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)

#evaluate our model
```
